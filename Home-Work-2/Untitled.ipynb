```python
import numpy as np
import matplotlib.pyplot as plt
```

# plot signal

### continuous time signal

```python
def plot_ct_signal(t, x_t):
    plt.grid()
    plt.xlim([t[0]-1, t[-1]+1])
    plt.plot(t, x_t)
```

### discrete time signal

```python
def plot_dt_signal(n, x_n):
    plt.grid()
    plt.xlim([n[0]-1, n[-1]+1])
    markerline, stemlines, baseline = plt.stem(n, x_n)
    plt.setp(baseline, color='k', linewidth=0.2)
```

```python
def U(a, b, i):
        if(a * i + b >= 0):
            return 1
        else:
            return 0
```

```python
k = np.arange(-25, 26, 1)
def h_n(i):
    return U(1, -10, i) - U(1, 10, i)
def x_n(i):
    return (U(1, -5, i) - U(1, 5, i)) * (0.25 ** i)
def y(j):
    s = 0
    for i in range(-50, 51):
        s = s + h_n(j - i) * x_n(i)
    return s

y_n = np.arange(-25, 26, 1)
for l in range(len(k)):
    y_n[l] = y(k[l])
plot_dt_signal(k, y_n)
```

```python
x = np.arange(-25, 26 , 1)
h = np.arange(-25, 26 , 1)
for i in range(len(h)):
    x[i] = x_n(x[i])
    h[i] = h_n(h[i])
y_n_c = np.convolve(x, h)
s = len(y_n_c) / 2 + 0.6
s = int(s)
plot_dt_signal(k, y_n_c[s-26:s+25])
```

## we see that 2 way to calculate convolution integral have same result

```python
k = np.arange(-15, 15.1, 0.1)
def h_t(i):
    return U(1, 3, i) - U(1, -3, i)
def x_t(i):
    return 0.5 * np.exp(2*i) * U(-1, 0, i)
def Y(j):
    s = 0
    i = -50
    while(i < 51):
        s = s + h_t(j - i) * x_t(i)
        i = i + 0.1
    return s

y_t = np.arange(-15, 15.1, 0.1)
for l in range(len(k)):
    y_t[l] = Y(k[l])
plot_ct_signal(k, y_t)
```

```python
x = np.arange(-15, 15.1 , 0.1)
h = np.arange(-15, 15.1 , 0.1)
for i in range(len(h)):
    x[i] = x_t(x[i])
    h[i] = h_t(h[i])

y_t_c = np.convolve(x, h)
s = len(y_t_c) / 2 + 0.6
s = int(s)
plot_ct_signal(k, y_t_c[s-151:s+150])
```

## we see that 2 way to calculate convolution sum have same result
