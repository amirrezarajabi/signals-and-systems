## Import Libraries

```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
def bk(k, data, T):
    
    w = np.pi * 2 / T
    s = 0
    for i in range(0, T):
        s = s + data[i]*np.cos(w*k*i)
    return s / T

def ck(k, data, T):
    
    w = np.pi * 2 / T
    s = 0
    for i in range(0, T):
        s = s + data[i]*np.sin(w*k*i)
    return s / T

def plotk(data, T):
    
    n = []
    x_n = []
    s = 0
    for i in range(-30, 31):
        d = (bk(i, data, T)**2+ck(i, data, T)**2)**0.5
        s = s + d
        if(11 > i > - 1):
            n.append(i)
            x_n.append(d)
    markerline, stemlines, baseline = plt.stem(n, x_n)
    plt.setp(baseline, color='k', linewidth=0.2)
    print("P:", s)
```

```python
def extract_wave(path, L):

    # Extract data
    samplerate, data = wavfile.read(path)

    data = np.reshape(data, (data.shape[0], 1))

    # Length
    length = data.shape[0] / samplerate

    time = np.linspace(0., length, data.shape[0])
    plt.plot(time[:L], data[:L, 0])
    plt.show()

    return (data, samplerate, length)
```

```python
data, samplerate, length = extract_wave("wave1.wav", 500)
```

```python
T = 100
plotk(data, T)
```
